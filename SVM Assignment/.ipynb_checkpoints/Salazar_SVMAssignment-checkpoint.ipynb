```python
import pandas as pd
data = pd.read_excel ('CTG.xls', 'Raw Data')
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.drop (['FileName', 'Date', 'SegFile', 'b', 'e', 'LBE'], axis=1)
```

```python
data.head()
```

```python
data = data.drop (['A', 'B', 'C', 'D', 'E','AD', 'DE', 'LD', 'FS', 'SUSP'], axis=1)
```

```python
data.head(10)
```

```python
data.info()
```

```python
data['NSP'].value_counts()
```

```python
data.isna().sum()
```

```python
data.dropna(axis = 0, inplace = True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

X = data.values[:, :-1]
y = data.values[:, -1]

x_train, x_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, random_state = 47)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

s = SVC(kernel='linear', gamma='auto')
s.fit (x_train, y_train)
y_pred_train = s.predict (x_train)
y_pred_test = s.predict (x_test)

print ('train accuracy=', accuracy_score (y_pred_train, y_train))
print ('test accuracy=', accuracy_score (y_pred_test, y_test))
```

```python
from sklearn.metrics import classification_report

print (classification_report (y_test, y_pred_test))
```
